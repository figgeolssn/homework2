```python
import numpy as np
import matplotlib.pyplot as plt

## TASK 1

class Interval:
    def __init__(self, left, right):
        """Initialize the Interval with left and right endpoints.

        Args:
        left (float): The left endpoint of the interval.
        right (float): The right endpoint of the interval.

        Raises:
        ValueError: If the left endpoint is greater than the right endpoint.
        """
        if left > right:
            raise ValueError("Left endpoint must not be greater than right endpoint.")
        
        self.left = left
        self.right = right

    def __repr__(self):
        """Return a string representation of the Interval."""
        return f"Interval({self.left}, {self.right})"
    
    def __str__(self):
        """Return a human-readable string representation of the Interval."""
        return f"[{self.left}, {self.right}]"

    def contains(self, number):
        """Check if the interval contains the given number.

        Args:
        number (float): The number to check.

        Returns:
        bool: True if the number is within the interval, False otherwise.
        """
        return self.left <= number <= self.right
    
    # Example Test
interval = Interval(1, 5)
print(interval)  # Output: [1, 5]
print(repr(interval))  # Output: Interval(1, 5)

# Check if a number is within the interval
print(interval.contains(3))  # Output: True
print(interval.contains(0))  # Output: False
print(interval.contains(5))  # Output: True
```

```python
## TASK 7

class Interval:
    def __init__(self, left, right=None):
        """
        Initialize the Interval with left and right endpoints.
        If only one argument is given, initialize a degenerate interval [left, left].

        Args:
            left (float): The left endpoint of the interval or the value of the degenerate interval.
            right (float, optional): The right endpoint of the interval. Defaults to None.

        Raises:
            ValueError: If the left endpoint is greater than the right endpoint.
        """
        if right is None:
            right = left  # If only one argument is provided, create a degenerate interval [left, left]
        
        if left > right:
            raise ValueError("Left endpoint must not be greater than right endpoint.")

        self.left = left
        self.right = right

    def __repr__(self):
        """Return a string representation of the Interval that could be used to recreate it."""
        return f"Interval({self.left}, {self.right})"

    def __str__(self):
        """Return a human-readable string representation of the Interval."""
        return f"[{self.left}, {self.right}]"

    def __add__(self, other):
        if not isinstance(other, Interval):
            raise ValueError("Operand must be an Interval.")
        return Interval(self.left + other.left, self.right + other.right)

    def __sub__(self, other):
        if not isinstance(other, Interval):
            raise ValueError("Operand must be an Interval.")
        return Interval(self.left - other.right, self.right - other.left)

    def __mul__(self, other):
        if not isinstance(other, Interval):
            raise ValueError("Operand must be an Interval.")
        products = [self.left * other.left, self.left * other.right,
                    self.right * other.left, self.right * other.right]
        return Interval(min(products), max(products))

    def __truediv__(self, other):
        if not isinstance(other, Interval):
            raise ValueError("Operand must be an Interval.")
        if other.left <= 0 <= other.right:
            raise ZeroDivisionError("Division by an interval containing zero is undefined.")
        divisions = [self.left / other.left, self.left / other.right,
                     self.right / other.left, self.right / other.right]
        return Interval(min(divisions), max(divisions))
    
# Create a degenerate interval
deg_interval = Interval(1)
print(deg_interval)  # Output: [1, 1]

# Create a normal interval
normal_interval = Interval(1, 5)
print(normal_interval)  # Output: [1, 5]
```

```python
## TASK 8

class Interval:
    def __init__(self, left, right=None):
        if right is None:
            right = left
        if left > right:
            raise ValueError("Left endpoint must not be greater than right endpoint.")
        self.left = left
        self.right = right

    def __repr__(self):
        return f"Interval({self.left}, {self.right})"

    def __str__(self):
        return f"[{self.left}, {self.right}]"

    def __add__(self, other):
        if isinstance(other, Interval):
            return Interval(self.left + other.left, self.right + other.right)
        else:  # Assume other is a scalar
            return Interval(self.left + other, self.right + other)

    def __radd__(self, other):
        return self.__add__(other)  # Addition is commutative

    def __sub__(self, other):
        if isinstance(other, Interval):
            return Interval(self.left - other.right, self.right - other.left)
        else:  # Assume other is a scalar
            return Interval(self.left - other, self.right - other)

    def __rsub__(self, other):
        return Interval(other - self.right, other - self.left)

    def __mul__(self, other):
        if isinstance(other, Interval):
            products = [self.left * other.left, self.left * other.right,
                        self.right * other.left, self.right * other.right]
            return Interval(min(products), max(products))
        else:  # Assume other is a scalar
            products = [self.left * other, self.right * other]
            if other >= 0:
                return Interval(min(products), max(products))
            else:
                return Interval(max(products), min(products))

    def __rmul__(self, other):
        return self.__mul__(other)  # Multiplication is commutative

    def __neg__(self):
        return Interval(-self.right, -self.left)

    def __truediv__(self, other):
        if isinstance(other, Interval):
            if other.left <= 0 <= other.right:
                raise ZeroDivisionError("Division by an interval containing zero is undefined.")
            divisions = [self.left / other.left, self.left / other.right,
                         self.right / other.left, self.right / other.right]
            return Interval(min(divisions), max(divisions))
        else:  # Assume other is a scalar
            if other == 0:
                raise ZeroDivisionError("Division by zero.")
            return Interval(self.left / other, self.right / other)

# Run the code from the task to test

print(Interval(2, 3) + 1)          # [3, 4]
print(1 + Interval(2, 3))          # [3, 4]
print(1.0 + Interval(2, 3))        # [3.0, 4.0]
print(Interval(2, 3) + 1.0)        # [3.0, 4.0]
print(1 - Interval(2, 3))          # [-2, -1]
print(Interval(2, 3) - 1)          # [1, 2]
print(1.0 - Interval(2, 3))        # [-2.0, -1.0]
print(Interval(2, 3) - 1.0)        # [1.0, 2.0]
print(Interval(2, 3) * 1)          # [2, 3]
print(1 * Interval(2, 3))          # [2, 3]
print(1.0 * Interval(2, 3))        # [2.0, 3.0]
print(Interval(2, 3) * 1.0)        # [2.0, 3.0]
print(-Interval(4, 5))             # see the special method __neg__
```

```python
class Interval:
    def __init__(self, left, right):
        self.left = left
        self.right = right

    def __str__(self):
        return f"[{self.left}, {self.right}]"

    def __pow__(self, n):
        if n % 2 == 0:
            left = min(self.left ** n, self.right ** n)
            right = max(self.left ** n, self.right ** n)
        else:
            left = min(self.left ** n, self.right ** n, (-self.left) ** n, (-self.right) ** n)
            right = max(self.left ** n, self.right ** n, (-self.left) ** n, (-self.right) ** n)
        return Interval(left, right)

# Example usage:
x = Interval(-2, 2)
print(x ** 2)  # Output: [4, 4]
print(x ** 3)  # Output: [-8, 8]
```

```python
## TASK 10

# Create lower and upper boundaries for 1000 intervals
xl = np.linspace(0.0, 1.0, 1000)
xu = xl + 0.5

class Interval:
    def __init__(self, left, right):
        self.left = left
        self.right = right

    def __pow__(self, exponent):
        return Interval(self.left**exponent, self.right**exponent)

    def __mul__(self, other):
        if isinstance(other, Interval):
            return Interval(self.left * other.left, self.right * other.right)
        else:  # Scalar multiplication
            return Interval(self.left * other, self.right * other)

    def __sub__(self, other):
        if isinstance(other, Interval):
            return Interval(self.left - other.right, self.right - other.left)
        else:  # Scalar subtraction
            return Interval(self.left - other, self.right - other)

    def __add__(self, other):
        if isinstance(other, Interval):
            return Interval(self.left + other.left, self.right + other.right)
        else:  # Scalar addition
            return Interval(self.left + other, self.right + other)

    def eval_polynomial(self):
        # p(I) = 3I^3 - 2I^2 - 5I - 1
        return Interval(3, 3) * (self ** 3) - Interval(2, 2) * (self ** 2) - Interval(5, 5) * self - Interval(1, 1)

# Create a list of interval objects
intervals = [Interval(xl[i], xu[i]) for i in range(1000)]

# Evaluate polynomial on each interval
results = [interval.eval_polynomial() for interval in intervals]

# Extract lower and upper boundaries
yl = [result.left for result in results]
yu = [result.right for result in results]

plt.figure(figsize=(10, 6))
plt.plot(xl, yl, label='Lower Boundary (yl)')
plt.plot(xl, yu, label='Upper Boundary (yu)')
plt.title('Polynomial Interval Evaluation')
plt.xlabel('Lower Boundary Values (xl)')
plt.ylabel('Boundary Values')
plt.legend()
plt.show()
```
